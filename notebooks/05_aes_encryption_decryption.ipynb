```python
# AES-256 Key Generation
import os
import secrets

KEY_DIR = "aes_keys"
os.makedirs(KEY_DIR, exist_ok=True)

KEY_PATH = f"{KEY_DIR}/aes_key.bin"

# Generate 256-bit AES key
key = secrets.token_bytes(32)

with open(KEY_PATH, "wb") as f:
    f.write(key)

print("üîê AES-256 key generated and saved.")
```

```python
import base64

with open(KEY_PATH, "rb") as f:
    key = f.read()

print("üîë AES key loaded successfully")
print("Key (Base64 preview):", base64.b64encode(key).decode()[:40], "...")
```

```python
from cryptography.hazmat.primitives.ciphers.aead import AESGCM
import secrets

def encrypt_image(input_path, encrypted_path, key):
    aesgcm = AESGCM(key)
    nonce = secrets.token_bytes(12)

    with open(input_path, "rb") as f:
        data = f.read()

    encrypted = aesgcm.encrypt(nonce, data, None)

    with open(encrypted_path, "wb") as f:
        f.write(nonce + encrypted)

    print(f"üîê Encrypted ‚Üí {encrypted_path}")
```

```python
from cryptography.hazmat.primitives.ciphers.aead import AESGCM

def decrypt_image(enc_path, output_img_path, key):
    aesgcm = AESGCM(key)

    with open(enc_path, "rb") as f:
        raw = f.read()

    nonce = raw[:12]
    encrypted_data = raw[12:]

    decrypted = aesgcm.decrypt(nonce, encrypted_data, None)

    with open(output_img_path, "wb") as f:
        f.write(decrypted)

    print(f"üîì Decrypted ‚Üí {output_img_path}")
    return decrypted
```
